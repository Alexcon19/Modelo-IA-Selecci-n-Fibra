```python
import csv

import requests
from bs4 import BeautifulSoup

# Lista para almacenar todas las citas, autores, etiquetas y fechas
all_quotes = []

# Número máximo de páginas a extraer (puedes ajustar esto según tus necesidades)
max_pages = 5

for page in range(1, max_pages + 1):
    url = f'https://quotes.toscrape.com/page/{page}/'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extracción y almacenamiento de citas, autores y etiquetas
    quotes = soup.find_all('span', class_='text')
    authors = soup.find_all('small', class_='author')
    tags = soup.find_all('div', class_='tags')

    for i in range(len(quotes)):
        quote_text = quotes[i].text
        author_name = authors[i].text
        tag_texts = [tag.text for tag in tags[i].find_all('a', class_='tag')]

        quote_info = {
            'Cita': quote_text,
            'Autor': author_name,
            'Etiquetas': tag_texts
        }

        all_quotes.append(quote_info)

# Imprimir todas las citas, autores y etiquetas almacenadas
for quote_info in all_quotes:
    print("Cita:", quote_info['Cita'])
    print("Autor:", quote_info['Autor'])
    print("Etiquetas:", quote_info['Etiquetas'])
    print("\n" + "-"*50 + "\n")

# Guardar la información en un archivo CSV
csv_file_path = 'resultados.csv'

with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    fieldnames = ['Cita', 'Autor', 'Etiquetas']
    csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    # Escribir encabezados
    csv_writer.writeheader()

    # Escribir datos
    for quote_info in all_quotes:
        csv_writer.writerow({
            'Cita': quote_info['Cita'],
            'Autor': quote_info['Autor'],
            'Etiquetas': ', '.join(quote_info['Etiquetas'])
        })

print(f"La información se ha guardado en {csv_file_path}")
```

```python
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
import matplotlib.pyplot as plt

# Cargar el archivo CSV (asegúrate de cambiar 'nombre_del_archivo.csv' por el nombre real de tu archivo)
df = pd.read_csv('resultados.csv')

# Combinar todas las citas en un solo texto
texto_completo = ' '.join(df['Cita'].astype(str))

# Tokenizar las palabras y convertirlas a minúsculas
palabras_tokenizadas = word_tokenize(texto_completo.lower())

# Eliminar palabras vacías (stopwords)
stop_words = set(stopwords.words('english'))  # Puedes ajustar el idioma según tus necesidades
palabras_filtradas = [palabra for palabra in palabras_tokenizadas if palabra.isalnum() and palabra not in stop_words]

# Contar la frecuencia de cada palabra
frecuencia_palabras = Counter(palabras_filtradas)

# Crear un DataFrame con la información
df_palabras = pd.DataFrame(list(frecuencia_palabras.items()), columns=['Palabra', 'Frecuencia'])

# Ordenar el DataFrame por frecuencia en orden descendente
df_palabras = df_palabras.sort_values(by='Frecuencia', ascending=False)

# Imprimir el DataFrame como tabla
print("Tabla de Frecuencia de Palabras:")
print(df_palabras)
print("\n" + "-"*50 + "\n")

# Obtener las 10 palabras más frecuentes para visualización
palabras_top = frecuencia_palabras.most_common(20)

# Visualizar las palabras más frecuentes en un gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(*zip(*palabras_top), color='skyblue')
plt.title('Palabras Más Frecuentes en las Citas')
plt.xlabel('Palabra')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```
